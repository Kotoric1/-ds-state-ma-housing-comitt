```python
# importing pandas library
import pandas as pd
# import matplotlib library
import matplotlib.pyplot as plt
# creating dataframe
#I manually typed this information by looking into the US census pulse survey website
#This should be automated to insure there are no miscalculations (did not add in final report)
df = pd.DataFrame({
    'Race': ['White', 'Black/African American', 'Asian', 'Hispanic/Latino'],
    'Phase 2 (8/19/20 to 10/26/20)': [10.77, 23.3, 11.7, 19.5],
    'Phase 3 (10/28/20 to 3/29/21)': [6.8, 21.6, 13.8, 20.5]
})
f = plt.figure()

plt.title('Percent of Massachussetts Who Missed a Rental or Mortgage Payment', color='black')
# plotting graph
ax = df.plot(x="Race", y=["Phase 2 (8/19/20 to 10/26/20)", "Phase 3 (10/28/20 to 3/29/21)"], kind="bar", ax=f.gca(), figsize=(10,6))
for p in ax.patches:
    ax.annotate(str(p.get_height()) + "%  ", (p.get_x() * 1.005, p.get_height() * 1.005))

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show() 
```
