```python
import pandas as pd
```

```python
gateway_cities =  [ 'Attleboro', 'Barnstable', 'Brockton', 'Chelsea', 'Chicopee', 'Everett', 'Fall River', 
                  'Fitchburg', 'Haverhill', 'Holyoke', 'Lawrence', 'Leominster', 
                  'Lowell', 'Lynn', 'Malden', 'Methuen','New Bedford', 'Peabody', 'Pittsfield', 'Quincy', 
                  'Revere', 'Salem', 'Springfield', 'Taunton', 'Westfield','Worcester']
```

```python
Gt10kLt50k = pd.read_csv('10kTO50K.csv') 
Gt50kLt90k = pd.read_csv('50kTO90K.csv')
Gt90kLt200k = pd.read_csv('90kTO200K.csv')
```

```python
normalized_df = pd.read_csv('household_normalization.csv')
```

```python
#Obtaining gateway cities and their corresponding per capita eviction rates
gc = normalized_df[normalized_df['Muni'].isin(gateway_cities)][['Muni','Evictions per Rented Households']]
```

```python
#Sorting gateway cities by descending order
gc.sort_values(by='Evictions per Rented Households',ascending=False,inplace=True)
```

```python
gc
```

```python
#Function that returns bracket level sorted gateway cities by per capita eviction rate
def gatewayCitiesByBracket(gc,df):
    return gc[gc.Muni.isin(df['Geographic Area Name'])]
```

```python
gatewayCitiesByBracket(gc,Gt10kLt50k)
```

```python
gatewayCitiesByBracket(gc,Gt50kLt90k)
```

```python
gatewayCitiesByBracket(gc,Gt90kLt200k)
```
